```python
# Loading the required packages and modules
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_excel("Superstore.xlsx")
df
```

```python
#  DATA VISULAISATION 
```

```python
# Group the data by the 'Category' column and calculate the total sales for each category
category_sales = df.groupby('Category')['Sales'].sum()

# Display the grouped data with total sales for each category
category_sales
```

```python
# Set the size of the figure to be displayed
plt.figure(figsize=(3, 2))

# Plot the data with specific settings
category_sales.plot(
    kind='line') # Specify the type of plot as a line plot
```

```python
plt.figure(figsize=(3, 2))

category_sales.plot(
    kind='line', 
    marker='o'  ) # Add circular markers at data points
```

```python
plt.figure(figsize=(3, 2))

category_sales.plot(
    kind='line',  
    marker='o',  
    color='black' # Set the line color to black
)
# Display the plot
plt.show()
```

```python
plt.figure(figsize=(3, 2))

category_sales.plot(
    kind='line',  
    marker='o',  
    color='black' 
)
# Add a label to the x-axis
plt.xlabel("Category of Super store")

# Add a label to the y-axis
plt.ylabel("Total Sales")

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(3, 2))

category_sales.plot(
    kind='line',  
    marker='o',  
    color='black'
)


plt.xlabel("Category of Super store")

plt.ylabel("Total Sales")

# Set the title of the plot
plt.title("Total sales by each category")

plt.show()
```

```python
plt.figure(figsize=(3, 2))

category_sales.plot(
    kind='line',  
    marker='o',  
    color='black'
)


plt.xlabel("Category of Super store")

plt.ylabel("Total Sales")

plt.title("Total sales by each category")

# Add grid lines to the plot
plt.grid(True, linestyle="--", alpha=1)  # Enable grid with dashed lines and full opacity

plt.show()
```

```python
# Group the data by the 'Region' column and calculate the total sales for each region
region_sales = df.groupby("Region")['Sales'].sum()

# Display the grouped data with total sales for each region
region_sales
```

```python
# Set the size of the figure to be displayed
plt.figure(figsize=(5, 4))

# Plot the data as a bar chart with specified colors
region_sales.plot(
    kind='bar',          # Specify the type of plot as a bar chart
    color=['red', 'blue', 'orange', 'green']  # Set specific colors for the bars
)

plt.xlabel("Region of Super store")

plt.ylabel("Total Sales")

plt.title("Total sales by each region")

plt.grid(True, linestyle="--", alpha=1)  

plt.show()
```

```python
# Set the size of the figure to be displayed
plt.figure(figsize=(5, 4))  # Properly close the parentheses here

# Create a scatter plot for 'Sales' and 'Profit' columns with specified transparency
plt.scatter(df['Sales'], df['Profit'], alpha=1)
           

plt.xlabel("total Sales")

plt.ylabel("Total Profits")

plt.title("Total Profits by Total Sales")

plt.grid(True, linestyle="--", alpha=1)  

plt.show()
```
